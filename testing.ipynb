```python
# Select numerical columns only to calculate correlations
numeric_df = t.select_dtypes(include=['number'])

# Calculate the correlation matrix
corr_matrix = numeric_df.corr()

# Find the top 12 features most correlated with 'RecommendationCount' (absolute value)
top_corr_features = corr_matrix['RecommendationCount'].abs().sort_values(ascending=False).head(12).index

# Plot the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(numeric_df[top_corr_features].corr(), annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Top 12 Features Correlated with RecommendationCount')
plt.show()
```

```python
t['DemoCount'].value_counts()   
```

```python
# 1. Check Distribution (Percentage of games with a Demo)
demo_percentage = (t['DemoCount'] > 0).mean() * 100
print(f"Percentage of games with at least 1 demo: {demo_percentage:.2f}%\n")

# 2. Check Correlation
correlation = t['DemoCount'].corr(t['RecommendationCount'])
print(f"Correlation between DemoCount and RecommendationCount: {correlation:.4f}\n")

# 3. Compare Means (Games with Demo vs No Demo)
t['HasDemo'] = (t['DemoCount'] > 0).astype(int)
print("Average RecommendationCount (0: No Demo, 1: Has Demo):")
display(t.groupby('HasDemo')['RecommendationCount'].mean())
```

```python
t['DLCCount'].value_counts()
```

```python
# 1. Categorize DLCCount: 0 for 0, 1 for 1, 2 for >1 (many)
t['DLCCountCat'] = np.where(t['DLCCount'] > 1, 2, t['DLCCount'])

# 2. Percentage of each count
print("Percentage of each DLC category (0: None, 1: One, 2: Multiple):")
dlc_percentages = t['DLCCountCat'].value_counts(normalize=True) * 100
print(dlc_percentages.sort_index().map('{:.2f}%'.format))
print("\n")

# 3. Correlation between the new DLCCountCat and RecommendationCount
correlation_dlc = t['DLCCountCat'].corr(t['RecommendationCount'])
print(f"Correlation between DLCCountCat and RecommendationCount: {correlation_dlc:.4f}\n")

# 4. Average Recs by Category
print("Average RecommendationCount by DLC Category:")
display(t.groupby('DLCCountCat')['RecommendationCount'].mean())
```

```python
# 1. Check raw correlation
print("Raw Correlation:", t['MovieCount'].corr(t['RecommendationCount']))

# 2. Check binary (Has Movie vs No Movie)
has_movie = (t['MovieCount'] > 0).astype(int)
print("Binary Correlation:", has_movie.corr(t['RecommendationCount']))

# 3. Check Combined Media
total_media = t['MovieCount'] + t['ScreenshotCount']
print("Combined Media Correlation:", total_media.corr(t['RecommendationCount']))

# 4. Check Capped MovieCount (e.g., maximum 5 movies)
capped_movies = t['MovieCount'].clip(upper=5)
print("Capped (Max 5) Correlation:", capped_movies.corr(t['RecommendationCount']))
```

```python
print((t['PriceFinal'] != t['PriceInitial']).sum())
print((t['PriceFinal'] < t['PriceInitial']).sum())
```

```python
# 1. Create a binary feature for discounts
t['HasDiscount'] = (t['PriceFinal'] < t['PriceInitial']).astype(int)

# 2. Check the percentage of games on discount
discount_pct = t['HasDiscount'].mean() * 100
print(f"Percentage of games currently on discount: {discount_pct:.2f}%\n")

# 3. Correlation between having a discount and popularity (RecommendationCount)
corr_discount = t['HasDiscount'].corr(t['RecommendationCount'])
print(f"Correlation between HasDiscount and RecommendationCount: {corr_discount:.4f}\n")

# 4. Average Recommendations for games with and without discounts
print("Average RecommendationCount by Discount Status (0: No Discount, 1: Has Discount):")
display(t.groupby('HasDiscount')['RecommendationCount'].mean())
```

```python
# 1. Check Distribution of PurchaseAvail
print("Distribution of PurchaseAvail:")
print(t['PurchaseAvail'].value_counts(normalize=True) * 100)
print("\n")

# 2. Convert boolean to integer for correlation
t['PurchaseAvail_Int'] = t['PurchaseAvail'].astype(int)
corr_purchase = t['PurchaseAvail_Int'].corr(t['RecommendationCount'])
print(f"Correlation between PurchaseAvail and RecommendationCount: {corr_purchase:.4f}\n")

# 3. Average Recommendations based on PurchaseAvail
print("Average RecommendationCount by PurchaseAvail:")
display(t.groupby('PurchaseAvail')['RecommendationCount'].mean())
```

```python
# Filter for rows where PurchaseAvail is False
no_purchase_avail = t[t['PurchaseAvail'] == False]

# Count how many of those have a Review that is not empty/null
# Using strip to make sure we don't count strings that are just spaces
valid_reviews = no_purchase_avail['Reviews'].astype(str).str.strip()
reviews_count = valid_reviews[valid_reviews != ''].count()

print(f"Total games where PurchaseAvail is False: {len(no_purchase_avail)}")
print(f"Number of those games that actually have a Review: {reviews_count}")
```

```python
category_columns = [
    'CategorySinglePlayer', 'CategoryMultiplayer', 'CategoryCoop', 
    'CategoryMMO', 'CategoryInAppPurchase', 'CategoryIncludeSrcSDK', 
    'CategoryIncludeLevelEditor', 'CategoryVRSupport'
]

# Calculate the average RecommendationCount for games where each category is True
avg_recs_by_category = {}
for col in category_columns:
    avg_recs = t[t[col] == True]['RecommendationCount'].mean()
    avg_recs_by_category[col] = avg_recs

# Convert to a Series and sort descending
category_popularity = pd.Series(avg_recs_by_category).sort_values(ascending=False)

print("Average RecommendationCount by Category (for games that have the category):")
display(category_popularity)

# Plot the results
plt.figure(figsize=(12, 6))
sns.barplot(x=category_popularity.values, y=category_popularity.index, palette='viridis')
plt.title('Average Popularity (RecommendationCount) by Game Category')
plt.xlabel('Average Recommendation Count')
plt.ylabel('Category')
plt.show()
```

```python
# 1. Identify all Genre columns in the dataset
genre_columns = [col for col in t.columns if 'Genre' in col]

if len(genre_columns) > 0:
    # 2. Calculate the average RecommendationCount for games where each genre is True
    avg_recs_by_genre = {}
    for col in genre_columns:
        # Assuming genres are boolean flags (True/False or 1/0)
        avg_recs = t[t[col] == True]['RecommendationCount'].mean()
        avg_recs_by_genre[col.replace('Genre', '')] = avg_recs # Remove 'Genre' prefix for cleaner labels

    # 3. Convert to a Series and sort descending
    genre_popularity = pd.Series(avg_recs_by_genre).dropna().sort_values(ascending=False)

    print("Average RecommendationCount by Genre (for games that have the genre):")
    display(genre_popularity)

    # 4. Plot the results
    plt.figure(figsize=(12, 8))
    sns.barplot(x=genre_popularity.values, y=genre_popularity.index, palette='magma')
    plt.title('Average Popularity (RecommendationCount) by Game Genre')
    plt.xlabel('Average Recommendation Count')
    plt.ylabel('Genre')
    plt.show()
else:
    print("Could not find columns containing 'Genre' (Check the exact column names using t.columns).")
```
