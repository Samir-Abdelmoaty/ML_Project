```python
""" 
remove nominal identfirs and text columns that are not useful for modeling and cannot be 
handled without text processing or embeddings
"""
text_and_id_cols = [
    'QueryID', 'QueryName', 'ResponseID', 'ResponseName', 'AboutText', 
    'Background', 'ShortDescrip', 'DetailedDescrip', 'DRMNotice', 
    'ExtUserAcctNotice', 'HeaderImage', 'LegalNotice', 'Reviews', 
    'PCMinReqsText', 'PCRecReqsText', 'LinuxMinReqsText', 
    'LinuxRecReqsText', 'MacMinReqsText', 'MacRecReqsText'
]
# Strip whitespace and convert to string type for all text and ID columns
for col in text_and_id_cols:
    t[col] = t[col].astype(str).str.strip()

t['ReleaseDate'] = pd.to_datetime(t['ReleaseDate'], errors='coerce')

reqs_cols = [
    'PCReqsHaveMin', 'PCReqsHaveRec', 'LinuxReqsHaveMin', 
    'LinuxReqsHaveRec', 'MacReqsHaveMin', 'MacReqsHaveRec'
]
# Preprocessing step: Convert True/False to 1 and 0
t[reqs_cols] = t[reqs_cols].astype(int)
```

```python
# 1. Select only numeric and boolean columns for correlation
numeric_df = t.select_dtypes(include=[np.number, bool])

# 2. Calculate correlation with RecommendationCount
correlations = numeric_df.corr()['RecommendationCount']

# 3. Get the top 10 features with the highest absolute correlation 
# (dropping RecommendationCount itself first)
top_10_features = correlations.drop('RecommendationCount').abs().sort_values(ascending=False).head(10).index

# Combine RecommendationCount with the top 10 features
features_to_plot = ['RecommendationCount'] + list(top_10_features)

# 4. Calculate the correlation matrix for these specific features
top_corr_matrix = t[features_to_plot].corr()

# 5. Create a mask for the upper triangle
mask = np.triu(np.ones_like(top_corr_matrix, dtype=bool))

# 6. Plot the heatmap using the mask
plt.figure(figsize=(10, 8))
sns.heatmap(top_corr_matrix, mask=mask, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5, square=True)
plt.title('Top 10 Features Correlated with RecommendationCount')
plt.show()
```
