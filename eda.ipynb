```python
from sklearn.preprocessing import StandardScaler

# Assuming RecommendationCount is our target for popularity
target = 'RecommendationCount'
num_cols = train_data.select_dtypes(include=np.number).columns.tolist()

if target in num_cols:
    # 1. Original Correlation
    orig_corr = train_data[num_cols].corr()[target]
    
    # 2. Standard Scaling (Linear Scaling)
    scaler = StandardScaler()
    scaled_data = pd.DataFrame(scaler.fit_transform(train_data[num_cols]), columns=num_cols)
    scaled_corr = scaled_data.corr()[target]
    
    # 3. Log Transformation (Non-linear Scaling)
    # Huge numbers in Steam datasets are often "power-law" distributed (highly skewed).
    # Using np.log1p safely applies log(x + 1) to handle zeros.
    log_data = train_data[num_cols].apply(lambda x: np.log1p(np.maximum(x, 0)))
    log_corr = log_data.corr()[target]
    
    # Combine into a DataFrame to compare the top 10 features
    top_features = orig_corr.abs().sort_values(ascending=False).index[1:11] # Skip the target itself
    
    comparison_df = pd.DataFrame({
        'Original Correlation': orig_corr[top_features],
        'Linear Scaled Corr': scaled_corr[top_features],
        'Log Transformed Corr': log_corr[top_features]
    })
    
    print("💡 OBSERVATION:")
    print("Standard (Linear) scaling DOES NOT change Pearson correlation. The formula ignores literal size.")
    print("However, Log transformation (Non-linear) changes the score and often reveals stronger, more stable relationships for large, skewed 'Steam' numbers.\n")
    
    display(comparison_df)
else:
    print(f"Target '{target}' not found in the dataset.")
```
